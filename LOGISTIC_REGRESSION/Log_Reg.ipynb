```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn import preprocessing
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('bank-full.csv',sep=';', header=0)
```

```python
df.head()
```

```python
df.describe()
```

```python
df = df.dropna()
```

```python
df.shape
```

```python
list(df.columns)
```

```python
df.drop(df.columns[[0, 3, 5, 8, 9, 10, 11, 12, 13,14]], axis=1, inplace=True)
```

```python
list(df.columns)
```

```python
data = pd.get_dummies(df, columns =['job', 'marital', 'default', 'housing','loan', 'poutcome'])
```

```python
data.head()
```

```python
data.drop(data.columns[[12,25]], axis=1, inplace=True)
```

```python
data.head()
```

```python
X = data.iloc[:,1:]
Y = data.iloc[:,0]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y,random_state=0)
```

```python
classifier = LogisticRegression(solver='lbfgs',random_state=0)
classifier.fit(X_train, Y_train)
```

```python
LogisticRegression(C=1.0, class_weight=None, dual=False,fit_intercept=True, intercept_scaling=1, max_iter=100, multi_class='warn', n_jobs=None, penalty='l2', random_state=0, solver='lbfgs',tol=0.0001, verbose=0, warm_start=False)
```

```python
predicted_y = classifier.predict(X_test)
```

```python
for x in range(len(predicted_y)):
    if (predicted_y[x] == 1):
        print(x, end="\t")
```

```python
print('Accuracy: {:.2f}'.format(classifier.score(X_test, Y_test)))
```
