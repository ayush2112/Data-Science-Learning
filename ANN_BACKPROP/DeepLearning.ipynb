```python
import numpy as np
import pandas as pd

observations = 1000
# generate random data
xs = np.random.uniform(low=-10,high=10,size=(observations,2))
zs = np.random.uniform(-10,10,(observations,2))
inputs = np.column_stack((xs,zs))
noise = np.random.uniform(-1,1,(observations,2))

targets = 2*xs - 3*zs + 5 + noise

```

```python
# initialize variables

init_range = 0.1
weights = np.random.uniform(-init_range,init_range,size=(4,2))
bias = np.random.uniform(-init_range,init_range,size = 2)
```

```python
# set learning rate
learning_rate = 0.02
```

```python
# train model
for i in range (100):
    output = np.dot(inputs,weights) + bias
    deltas = output - targets
    loss = np.sum(deltas**2)/2/observations  #loss per observation
    print(loss)
    deltas_scaled = deltas/observations
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    bias = bias - learning_rate * np.sum(deltas_scaled)
    print(weights,bias)
```
